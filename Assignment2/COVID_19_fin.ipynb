##Field description
* **Province/State**: China - province name; US/Canada/Australia/ - city name, state/province name; Others - name of the event (e.g., "Diamond Princess" cruise ship); other countries - blank.
* **Country/Region**: country/region name conforming to WHO (will be updated).
* **Confirmed**: the number of confirmed cases(accumulated).
* **Deaths**: the number of deaths(accumulated).
* **Recovered**: the number of recovered cases(accumulated).

```
import pandas as pd
import pycountry_convert as pc
cf = pd.read_csv('time_series_covid19_confirmed_global.csv', delimiter = ',')
dt = pd.read_csv('time_series_covid19_deaths_global.csv', delimiter = ',')
rc = pd.read_csv('time_series_covid19_recovered_global.csv', delimiter = ',')
```

```
cf['Country/Region'] = cf['Country/Region'].replace(
                            ['Congo (Brazzaville)', 'Congo (Kinshasa)', "Cote d'Ivoire", 'Burma', 'Korea, South', 'Taiwan*', 'US'], 
                            ['Congo', 'Congo', 'Ivory Coast', 'Myanmar', 'South Korea', 'Taiwan', 'United States']
                            )

rc['Country/Region'] = rc['Country/Region'].replace(
                            ['Congo (Brazzaville)', 'Congo (Kinshasa)', "Cote d'Ivoire", 'Burma', 'Korea, South', 'Taiwan*', 'US'], 
                            ['Congo', 'Congo', 'Ivory Coast', 'Myanmar', 'South Korea', 'Taiwan', 'United States']
                            )

dt['Country/Region'] = dt['Country/Region'].replace(
                            ['Congo (Brazzaville)', 'Congo (Kinshasa)', "Cote d'Ivoire", 'Burma', 'Korea, South', 'Taiwan*', 'US'], 
                            ['Congo', 'Congo', 'Ivory Coast', 'Myanmar', 'South Korea', 'Taiwan', 'United States']
                            )
```

```
country_list = cf['Country/Region'].unique()
len(country_list)
```

```
convert = {}

for country in country_list:
    try :
        convert[country] = pc.convert_continent_code_to_continent_name(pc.country_alpha2_to_continent_code(pc.country_name_to_country_alpha2(country)))
    except :
        convert[country] = "Other"

len(convert)
```

```
cf['Continents'] = cf['Country/Region'].map(convert)
dt['Continents'] = dt['Country/Region'].map(convert)
rc['Continents'] = rc['Country/Region'].map(convert)
```

```
rc_result = rc.drop(axis=1, columns=['Province/State', 'Lat', 'Long'])
rc_max = rc_result.groupby(['Country/Region']).agg('sum')
rc_max = rc_max['4/12/20'].to_frame('Max_Recover')
cf_result = cf.drop(axis=1, columns=['Province/State', 'Lat', 'Long'])
cf_max = cf_result.groupby(['Country/Region']).agg('sum')
cf_max = cf_max['4/12/20'].to_frame('Max_Confirmed')
dt_result = dt.drop(axis=1, columns=['Province/State', 'Lat', 'Long'])
dt_max = dt_result.groupby(['Country/Region']).agg('sum')
dt_max = dt_max['4/12/20'].to_frame('Max_Death')
```

```
eu_list = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy','Latvia', 'Lithuanai', 'Luxenbourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden']
x = rc_max.reset_index()
rc_eu = x[x['Country/Region'].isin(eu_list)]
y = cf_max.reset_index()
cf_eu = y[y['Country/Region'].isin(eu_list)]

eu = pd.merge(rc_eu, cf_eu, how='inner', left_on='Country/Region', right_on='Country/Region')
eu['Union'] ='European Union'
eu = eu.groupby(['Union']).agg('sum')
eu['Percent_recovered'] = eu['Max_Recover'] / eu['Max_Confirmed'] * 100

print(eu['Percent_recovered'])
print ("\nอัตรการรักษาหายของผู้ป่วยในสหภาพยุโรปเท่ากับ " + " " + str(eu.loc['European Union', 'Percent_recovered']) + "%")
```

##2.   จงหา Country/Region ที่มียอดจำนวนผู้ติดเชื้อล่าสุดมากที่สุด 10 อันดับพร้อมพล็อตกราฟแสดงจำนวนผู้ติดเชื้อของแต่ละ Country/Region โดยเรียงลำดับจากมากไปน้อย (ไม่ต้องแยกเป็น Province/State)

```
#Solution
top10_cf = cf_result.groupby(['Country/Region']).agg('sum')
top10_cf = top10_cf.nlargest(10, '4/12/20')
top10_cf = top10_cf['4/12/20'].to_frame('Max_Confirmed')

print('Country/Region ที่มียอดจำนวนผู้ติดเชื้อล่าสุดมากที่สุด 10 อันดับ\n')
print(top10_cf)
print("\n")

top10_cf.plot(kind='bar', subplots=True, figsize=(5, 5), legend=True, grid=True)
```

##3.   จงพล็อตกราฟเส้นเพื่อแสดง trend ยอดผู้ติดเชื้อในประเทศไทยในแต่ละสัปดาห์(กำหนดให้นับข้อมูลทุกวันอาทิตย์)

```
#Solution
thai = cf_result.groupby('Country/Region').agg('sum')
thai = thai.T['Thailand'].to_frame('Thailand')
thai = thai.reset_index()
datetime = pd.to_datetime(thai['index'])
dayname = datetime.dt.day_name()
thai['Day_Name'] = dayname

#print(thai)
#print("\n")
thai_sunday = thai[thai['Day_Name'] == 'Sunday']

#print(thai_sunday)
#print("\n")
print('กราฟเส้นแสดง trend ยอดผู้ติดเชื้อในประเทศไทยแต่ละสัปดาห์ (ทุกวันอาทิตย์)\n')

line_chart = thai_sunday.drop(axis=1, columns=['Day_Name'])
line_chart.plot.line(x='index', y='Thailand', figsize=(10, 5), legend=True, grid=True, linestyle='-', marker='o')
```

##4.   จงหายอดผู้ติดเชื้อของ**แต่ละวัน**บนเรือไดม่อนพริ้นเซสพร้อมพล็อตกราฟ แล้วระบุวันที่มีการติดเชื้อมากที่สุด

```
#Solution
diamond = cf_result.groupby('Country/Region').agg('sum')
diamond = diamond.T['Diamond Princess'].to_frame('Diamond_Princess')

#print(diamond)
#print('\n')


plot_diamond = diamond.diff()
plot_diamond['Diamond_Princess'] = plot_diamond['Diamond_Princess'].replace([None], [0])

print('ยอดผู้ติดเชื้อของแต่ละวันบนเรือไดม่อนพริ้นเซส\n')
print(plot_diamond)
print('\n')
print('กราฟแสดงยอดผู้ติดเชื้อของแต่ละวันบนเรือไดม่อนพริ้นเซส\n')

plot_diamond.plot(kind='line', figsize=(10, 5), legend=True, grid=True, linestyle='-', marker='o')
```

```
plot_diamond = plot_diamond.idxmax(axis=0, skipna=True)
print('วันที่มีการติดเชื้อมากที่สุด คือ วันที่ ' + plot_diamond)
```

##5.   จงหา Country/Region ที่ล่าสุดมีเปอร์เซ็นต์อัตราการเสียชีวิตมากที่สุด 20 อันดับ พร้อมระบุว่า จาก Country/Region ในกลุ่มดังกล่าว ส่วนใหญ่อยู่ในทวีปอะไร


```
#Solution
max_dt = dt_result.groupby(['Country/Region']).agg('sum')
max_cf = cf_result.groupby(['Country/Region']).agg('sum')

mdt = max_dt['4/12/20'].to_frame('Max_Death')
mcf = max_cf['4/12/20'].to_frame('Max_Confirmed')

mdt['Max_Confirmed'] = mcf['Max_Confirmed']

#print(mdt)
#print('\n')
print('Country/Region ที่มีอัตราเสียชีวิตมากที่สุด 20 อันดับ\n')

mdt['Percent_Death'] = mdt['Max_Death'] / mdt['Max_Confirmed'] * 100

mdt = mdt.reset_index()
mdt['Continents'] = mdt['Country/Region'].map(convert)
mdt

top20_dt = mdt.nlargest(20, 'Percent_Death')
print(top20_dt[['Country/Region', 'Continents', 'Percent_Death']])
```

```
count_continents = top20_dt.pivot_table(index=['Continents'], aggfunc='size')
count_continents = count_continents.idxmax(axis=0, skipna=True)
print('ทวีปที่มีอัตราการเสียชีวิตมากที่สุด คือ ' + count_continents)
```

##6.   จงหายอดผู้ป่วยที่กำลังรักษาตัวล่าสุด (สมมติว่ายังไม่เสียชีวิตหรือหายดี) ของแต่ละ Province/State ในประเทศจีน พร้อมพล็อตกราฟแสดงจำนวนในแต่ละ Province/State

```
#Solution
chinarc = rc.groupby(['Country/Region', 'Province/State'])[['Country/Region', 'Province/State', '4/12/20']].agg('sum')
chinarc = chinarc.reset_index()
chinarc = chinarc[(chinarc['Country/Region'] == 'China')][["Province/State", 'Country/Region', '4/12/20']]


chinacf = cf.groupby(['Country/Region', 'Province/State'])[['Country/Region', 'Province/State', '4/12/20']].agg('sum')
chinacf = chinacf.reset_index()
chinacf = chinacf[(chinacf['Country/Region'] == 'China')][["Province/State", 'Country/Region', '4/12/20']]


chinadt = dt.groupby(['Country/Region', 'Province/State'])[['Country/Region', 'Province/State', '4/12/20']].agg('sum')
chinadt = chinadt.reset_index()
chinadt = chinadt[(chinadt['Country/Region'] == 'China')][["Province/State", 'Country/Region', '4/12/20']]


#chinarc.shape
#chinacf.shape
#chinadt.shape

chinarc = chinarc.drop(axis=1, columns=['Country/Region'])
chinadt = chinadt.drop(axis=1, columns=['Country/Region'])

china = pd.merge(chinacf, chinarc, how='inner', left_on='Province/State', right_on='Province/State')
china = pd.merge(china, chinadt, how='inner', left_on='Province/State', right_on='Province/State')
china.rename(columns={'4/12/20_x' : 'Confirmed','4/12/20_y' : 'Recovered', '4/12/20' : 'Death'}, inplace=True)
china['In_Process'] = china['Confirmed'] - china['Recovered'] - china['Death']
china_inprocess = china[['Province/State', 'In_Process']]

print('ยอดผู้ป่วยที่กำลังรักษาตัวล่าสุด ของแต่ละ Province/State ในประเทศจีน\n')
print(china_inprocess)
print('\n')

china_inprocess.plot.bar(x='Province/State', y='In_Process', figsize=(15, 5), legend=True, grid=True)
```
