##3) เขียนโปรแกรมเพื่อหายอดขายรวมของสินค้าทุกประเภท ในทุกช่องทาง สำหรับแต่ละภูมิภาคแล้ววาด pie chart

###Solution 1

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')
```

```
#sum by row then store in new column
wholesale['Total'] = wholesale.sum(axis=1, skipna=True)

wholesale.head(10)
```

```
#get each region name
region = wholesale.Region.unique()
```

```
#use dictionary for result
result = {}

#sum last column in each region
for reg in region:
  result[reg] = wholesale.Total[wholesale.Region == reg].sum(axis=0, skipna=True)

result
```

```
#create dataframe from dictionary
pie_chart = pd.DataFrame.from_dict(result, orient='index', columns=['Regions'])

#plot directly from dataframe
pie_chart.plot.pie(subplots=True, figsize=(10, 10), autopct='%.2f%%', startangle=90, counterclock=False, pctdistance=1.1, labeldistance=1.2)
```

###Solution 2

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')
result = wholesale.groupby(['Region']).agg('sum')

print(result)

result = result.sum(axis=1, skipna=True).sort_values(ascending=False).to_frame(name='Regions')


result.plot.pie(subplots=True, figsize=(10, 10), autopct='%.2f%%', startangle=90, counterclock=False, pctdistance=1.1, labeldistance=1.2)
```

##4) เขียนโปรแกรมเพื่อหายอดขายรวมของสินค้าสำหรับทำความสะอาดในแต่ละช่องทางการขาย โดยแบ่งเป็นของแต่ละภูมิภาค พร้อมพล็อตกราฟของแต่ละภูมิภาค

###Solution 1

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')
```

```
all_channel = sorted(wholesale.Channel.unique())
all_region = sorted(wholesale.Region.unique())

print("Channel list --> ",all_channel,"\n","Region list -->",region)
```

```
each_channel = {}

for reg in all_region:
  temp = []
  for cha in all_channel:
    temp.append(wholesale.Detergents_Paper[(wholesale.Region == reg) & (wholesale.Channel == cha)].sum(axis=0, skipna=True))
  each_channel[reg] = temp

each_channel
```

```
#create dataframe from dictionary
pie_chart = pd.DataFrame.from_dict(each_channel, orient='columns')
```

```
#plot directly from dataframe
pie_chart.plot.pie(subplots=True, labels=all_channel, figsize=(20, 20), autopct='%.2f%%', startangle=90, counterclock=False, pctdistance=1.1, labeldistance=1.2)
```

###Solution 2

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')

result = wholesale.groupby(['Channel','Region']).agg('sum')
detergents_paper = result['Detergents_Paper']

print(detergents_paper)

detergents_paper.to_frame().unstack(level=1).plot(kind='bar', subplots=True, figsize=(10, 15), legend=True, grid=True)
```

##5) เขียนโปรแกรมเพื่อหาจำนวนปีที่ยอดขายของอาหารสดรวมกับอาหารแช่แข็งผ่านช่องทาง Horeca(Hotel/Restaurant/Cafe) ไม่เกิน 1,000 ชิ้น

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')
```

```
wholesale['Fresh_and_frozen'] = wholesale['Fresh'] + wholesale['Frozen']

wholesale.head(10)
```

```
horeca_fresh_frozen = wholesale[(wholesale.Fresh_and_frozen <= 1000) & (wholesale.Channel == 'Horeca')]

print('จำนวน',horeca_fresh_frozen.shape[0],'ปี')
```

##6) เขียนโปรแกรมเพื่อหาภูมิภาคที่มีการบริโภคนมมากที่สุด และในภูมิภาคนั้นมีการบริโภคนมมากกว่าอาหารสดทั้งหมดกี่ปี

```
wholesale = pd.read_csv('Rename_Wholesale.csv', delimiter=',')
```

```
sum_region = wholesale.groupby(['Region']).sum()
sum_region
```

```
max_milk_region = sum_region['Milk'].idxmax(axis=0, skipna=True) 
print("Region ที่บริโภคนมมากสุดคือ", max_milk_region)
```

```
milk_over_fresh = wholesale[(wholesale.Milk > wholesale.Fresh) & (wholesale.Region == max_milk_region)]
print('บริโภคนมมากกว่าอาหารสดจำนวน',milk_over_fresh.shape[0],'ปี')
```
